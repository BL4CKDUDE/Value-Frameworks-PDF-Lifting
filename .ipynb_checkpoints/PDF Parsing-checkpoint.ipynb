```python
def extract_text_and_tables(pdf_path):
    text_content = ""
    tables = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Extract text content from the page
            text_content += page.extract_text()

            # Extract tables from the page
            page_tables = page.extract_tables(table_settings={"vertical_strategy": "text", 
                                               "horizontal_strategy": "text"})
            if page_tables:
                tables.extend(page_tables)

    return text_content, tables

# Example usage
pdf_path = repo_path+"/Input Data/Data Sources/Structured data/2022-Absa-Group-limited-Environmental-Social-and-Governance-Data-sheet.pdf"
text_content, tables = extract_text_and_tables(pdf_path)
```
