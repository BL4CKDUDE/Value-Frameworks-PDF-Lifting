```python
df_pdf_names = pd.read_csv(repo_path+"/Input Data/PDFnames.csv", )
tk_pandas_essentials.desc_df(df_pdf_names)
```

```python
pdfs = []
for i,row in df_pdf_names.iterrows():
    pdf_filepath = pdf_path + "/" + str(row["File Name"])
    pdfs.append(pdf_filepath)
len(pdfs)
```

```python
filename = "2022-Absa-Group-limited-Environmental-Social-and-Governance-Data-sheet.pdf"
index = 0
training_sets = []
for pdf_fullpath in pdfs:
    if filename.lower() in pdf_fullpath.lower():
        doc_id = df_pdf_names["ID"].iloc[index]
        doc_name = df_pdf_names["File Name"].iloc[index]
        group = df_pdf_names["ID"].iloc[index].split(".")[0]
        training_data = tk_pdf_parser.tabulate(pdf_fullpath, doc_id, doc_name, group)
        training_sets.append(training_data)
        
        # print(doc_id)
        # print(doc_name)
        # print(group)
        # print(index)
        # tk_pandas_essentials.stop()
    index = index + 1
```
