```python
pdfs = []
for i,row in df_pdf_names.iterrows():
    pdf_filepath = pdf_path + "/" + str(row["File Name"])
    pdfs.append(pdf_filepath)
len(pdfs)
```

```python
def extract_tables_from_pdf(pdf_path):
    tables = []
    for page_layout in extract_pages(pdf_path):
        for element in page_layout:
            if isinstance(element, LTTextBox) or isinstance(element, LTTextContainer):
                text = element.get_text().strip()
                if any(row.strip() for row in text.split('\n')):
                    rows = [row.strip().split('\n') for row in text.split('\n\n')]
                    if all(len(row) > 1 for row in rows):
                        tables.append(rows)
    return tables

filename = "SASOL Sustainability Report 2023 20-09_0.pdf"
index = 0
training_sets = []
for pdf_fullpath in pdfs:
    if filename.lower() in pdf_fullpath.lower():
        doc_id = df_pdf_names["ID"].iloc[index]
        doc_name = df_pdf_names["File Name"].iloc[index]
        group = df_pdf_names["ID"].iloc[index].split(".")[0]
        training_data = extract_tables_from_pdf(pdf_fullpath, doc_id, doc_name, group)
        training_sets.append(training_data)
        
    index = index + 1
```
